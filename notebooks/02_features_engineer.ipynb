```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# Plot distributions for all engineered features
sns.histplot(data=df, x='hour', hue='isFraud', bins=24, ax=axes[0, 0])
axes[0, 0].set_title('Hour Distribution')

sns.countplot(data=df, x='is_night', hue='isFraud', ax=axes[0, 1])
axes[0, 1].set_title('Night Activity (0=day, 1=night)')

sns.countplot(data=df, x='is_high_risk_type', hue='isFraud', ax=axes[0, 2])
axes[0, 2].set_title('High-Risk Transaction Type')

sns.kdeplot(data=df, x='log_amount', hue='isFraud', ax=axes[1, 0], common_norm=False)
axes[1, 0].set_title('Log Amount Distribution')

sns.countplot(data=df, x='is_dest_zero_balance', hue='isFraud', ax=axes[1, 1])
axes[1, 1].set_title('Destination Zero Balance')

sns.countplot(data=df, x='high_risk_combo', hue='isFraud', ax=axes[1, 2])
axes[1, 2].set_title('High-Risk Combo')

plt.tight_layout()
plt.show()
```
