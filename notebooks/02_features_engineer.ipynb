```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../')

# Import modules from folder src
from src.data_loader import load_paysim_data
from src.features import build_features

# Load raw data
df = load_paysim_data('D:/Công việc/DA + DE/Book DA/Data Science for Business/PaySim/fraud-detection-paysim/data/raw/Synthetic_Financial_datasets_log.csv')

# Processing Feature Enginerring balance function was packaged
df = build_features(df)

# Check data after changed
df.info()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(14, 10))

# 1. Hour distribution by fraud
sns.histplot(data=df, x='hour', hue='isFraud', bins=24, ax=axes[0, 0])
axes[0, 0].set_title('Hour Distribution by Fraud Status')

# 2. Night vs Fraud
sns.countplot(data=df, x='is_night', hue='isFraud', ax=axes[0, 1])
axes[0, 1].set_title('Night Activity vs Fraud')

# 3. High-risk type distribution
sns.countplot(data=df, x='is_high_risk_type', hue='isFraud', ax=axes[1, 0])
axes[1, 0].set_title('High-Risk Transaction Type vs Fraud')

# 4. Log amount distribution
sns.kdeplot(data=df, x='log_amount', hue='isFraud', ax=axes[1, 1], common_norm=False)
axes[1, 1].set_title('Log Amount Distribution by Fraud')

# 5. Destination zero balance
sns.countplot(data=df, x='is_dest_zero_balance', hue='isFraud', ax=axes[2, 0])
axes[2, 0].set_title('Destination Zero Balance vs Fraud')

# 6. High-risk combo (type + night + zero balance)
sns.countplot(data=df, x='high_risk_combo', hue='isFraud', ax=axes[2, 1])
axes[2, 1].set_title('High-Risk Combo vs Fraud')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df[['amount', 'log_amount', 'hour', 'is_night', 'is_high_risk_type', 
                          'is_dest_zero_balance', 'high_risk_combo', 'isFraud']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', 
            square=True, linewidths=1)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Fraud rate by feature
features_to_analyze = [
    ('is_high_risk_type', 'Fraud Rate by Risk Type'),
    ('is_night', 'Fraud Rate by Time of Day'),
    ('is_dest_zero_balance', 'Fraud Rate by Destination Balance'),
    ('high_risk_combo', 'Fraud Rate by High-Risk Combo')
]

for idx, (feature, title) in enumerate(features_to_analyze):
    ax = axes[idx // 2, idx % 2]
    fraud_rate = df.groupby(feature)['isFraud'].agg(['sum', 'count'])
    fraud_rate['rate'] = fraud_rate['sum'] / fraud_rate['count'] * 100
    
    fraud_rate['rate'].plot(kind='bar', ax=ax, color=['green', 'red'][:len(fraud_rate)])
    ax.set_title(title)
    ax.set_ylabel('Fraud Rate (%)')
    ax.set_xlabel(feature)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
    
    # Add percentage labels on bars
    for i, v in enumerate(fraud_rate['rate']):
        ax.text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
